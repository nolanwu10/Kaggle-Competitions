```python
import pandas as pd
```

```python
df= pd.read_csv('/content/drive/MyDrive/Machine Learning/Data/train (2).csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import median_absolute_error
```

```python
features = df.drop(columns=['id', 'Hardness'])
target = df['Hardness']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
mae = median_absolute_error(y_test, y_pred)
mae
```

```python
test = pd.read_csv('/content/drive/MyDrive/Machine Learning/Data/test (3).csv')
```

```python
test_features = test.drop(columns=['id'])

test_predictions = model.predict(test_features)
submission = pd.DataFrame({
    'id': test['id'],
    'Hardness': test_predictions
})
submission.head()
```

```python
submission.to_csv('/content/drive/MyDrive/Machine Learning/intro to machine learning/kaggle1130.csv', index = False)
```

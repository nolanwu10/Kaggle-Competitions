```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Machine Learning/Data/housing/train (3).csv')
```

```python
df.head()
```

```python
df.isna().sum().reset_index()
```

```python
features = ['Id', 'MSSubClass', 'LotArea', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold']
```

```python
X = df[features]
```

```python
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state =1)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

y_pred = rf_model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))

rmse
```

```python
testfeatures = ['Id', 'MSSubClass', 'LotArea', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold']
```

```python
test_data = pd.read_csv('/content/drive/MyDrive/Machine Learning/Data/housing/test.csv')


X_test_final = test_data[testfeatures]

predicted_sale_prices = rf_model.predict(X_test_final)


submission = pd.DataFrame({
    'Id': test_data['Id'],
    'SalePrice': predicted_sale_prices
})


submission.to_csv('/content/drive/MyDrive/Machine Learning/Data/housing/submission.csv', index=False)

# This line is just to confirm that the submission DataFrame has been created correctly
print(submission.head())
```
